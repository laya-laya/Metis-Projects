# Predict average dollar value in auction drafts
At the start of a fantasy sports season, each participant needs to draft players to their fantasy team. One way to do this is an auction style draft, where each player put up for auction goes to the highest bidder. Usually a set amount of "fantasy dollars" are given to each team, and they have to budget accordingly.

Yahoo provides some data on the average fantasy dollar value spent on each player in auction style drafts. We will try to predict the average dollar value spent on each player given the previous years stats. This prediction is useful for people looking for insights into auction drafts.

```python
from bs4 import BeautifulSoup
from collections import defaultdict

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn import metrics

from ngboost import NGBRegressor
from ngboost.distns import Exponential, Normal, LogNormal
from ngboost.scores import LogScore, CRPS

import warnings
warnings.filterwarnings("ignore")
```

### Loading data
Scrape from downloaded HTML pages

```python
# open html data files
player_pages = []
for page_idx in range(1,6):
    with open("data/auction/auction{}.html".format(page_idx), "r") as f:
        player_pages.append(f.read())

            
# convert files to soup
player_soups = [BeautifulSoup(player_pages[idx], "html.parser") for idx in range(5)]


def parse_name_row(name_str):
    split_idx = name_str.rfind("-")
    name_team, pos = (name_str[:split_idx].strip(), name_str[split_idx+1:].strip())
    names_and_team = name_team.split(' ')
    team = names_and_team.pop()
    name = " ".join(names_and_team).strip() # team name already pop'd out
    return name, team.strip(), pos


def get_df(soups):
    """
    create a dataframe from beautifulsoup objects provided. converts features to ints when possible
    """
    columns = defaultdict(list)
    for soup in soups:
        rows = soup.find(class_="Table Ta-start Fz-xxs Table-interactive ysf-scrollloader-table").tbody.find_all("tr")
        for row in rows:
            player = dict()
            ds = row.find_all("td")
            
            name, team, position = parse_name_row(ds[0].text.split("\n")[2])
            columns["name"].append(name)
            columns["team"].append(team)
            columns["position"].append(position)
            
            columns["avg_value"].append(float(ds[2].text.replace("$", '')))
    return pd.DataFrame(columns)


avg_value = get_df(player_soups)
```

```python
# open html data files
player_pages = defaultdict(list)
seasons = [17, 18, 19]
for season in seasons:
    for page_idx in range(1,8):
        with open("data/{}/players{}.html".format(season, page_idx), "r") as f:
            player_pages[season].append(f.read())

            
# convert files to soup
player_soups = dict()
for season in seasons:
    player_soups[season] = [BeautifulSoup(player_pages[season][idx], "html.parser") for idx in range(7)]

    
def secs(time_str):
    mins, secs = map(int, time_str.split(':'))
    seconds = mins*60 + secs
    return seconds


def parse_name_row(name_str):
    split_idx = name_str.rfind("-")
    name_team, pos = (name_str[:split_idx].strip(), name_str[split_idx+1:].strip())
    names_and_team = name_team.split(' ')
    team = names_and_team.pop()
    name = " ".join(names_and_team).strip() # team name already pop'd out
    return name, team.strip(), pos


def get_season_df(soups, season):
    """
    create a dataframe from beautifulsoup objects provided. converts features to ints when possible
    """
    columns = defaultdict(list)
    for soup in soups:
        rows = soup.find(class_="Table Ta-start Fz-xs Table-mid Table-px-xs Table-interactive").tbody.find_all("tr")
        for row in rows:
            player = dict()
            ds = row.find_all("td")
            
            columns["season"].append(season)
            
            name, team, position = parse_name_row(ds[1].text.split("\n")[2])
            columns["name"].append(name)
            columns["team"].append(team)
            columns["position"].append(position)
            
            columns["gp"].append( int(ds[5].text) )  # games played
            columns["rank"].append( int(ds[7].text) )  # player rank
            columns["toi"].append( secs(ds[9].text) )  # time on ice
            columns["goals"].append( int(ds[10].text) )  # goals
            columns["assists"].append( int(ds[11].text) )  # assists
            columns["points"].append( int(ds[12].text) )  # total points
            columns["+/-"].append( int(ds[13].text) )  # plus minus
            columns["ppp"].append( int(ds[14].text) )  # power play points
            columns["sog"].append( int(ds[15].text) )  # shots on goal
            columns["hits"].append( int(ds[16].text) )  # hits
            columns["blocks"].append( int(ds[17].text) )  # blocks
    
    return pd.DataFrame(columns)


season_dfs = []
for season in seasons:
    season_dfs.append(get_season_df(player_soups[season], season))

players = pd.concat(season_dfs)
```

```python
players_18 = players[players["season"] == 18]
players_avg_val = players_18.merge(avg_value)
```

## Split the data

```python
# train test split
X = players_avg_val.drop(columns=["season", "team", "points", "position", "avg_value",])
                                 #"season_17", "team_17", "points_17", "position_17"])
y = players_avg_val["avg_value"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.15, random_state=2)

# get names
X_train_names = X_train["name"]
X_train = X_train.drop(columns="name").values.astype(float)

X_test_names = X_test["name"]
X_test = X_test.drop(columns="name").values.astype(float)

# just the ranks
X_train_ranks = X_train[:,1]
X_test_ranks = X_test[:,1]

# scaling features
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Modeling 1 - simple model
First I will try to fit a simple inverse exponential curve to this data using gradient descent.

f = 60 / (e^(x/w))        -- where x is the player rank, and w is a weight, or parameter. 60 was chosen to be the numerator because when x (player rank) is 1, the function returns a little less than 60, which is about equal to the average value of the first rank player in this data.

df/dw = 60x / (w^2 * e^(x/w))      -- derivative of f wrt w

MSE = (1/n) * sum([(f(x_i) - y_i)^2 for i in range(len(y))])   -- mean squared error (loss function)

dMSE/dw = (1/n) * sum([2(f(x_i) - y_i) * 60x_i/(w^2 * e^(x_i / w)])

```python
# the inverse exponential function seen in the graph above (green line)
def f(x, w):
    return 60/(np.exp(x/w))
```

```python
# MSE for a single data point
def mse(y_hat, y):
    return (y_hat - y)**2
```

```python
# derivative of MSE of inverse exponential function wrt w, for a single data point
def d_mse_wrt_w(x, w, y):
    pred = f(x, w)
    return (pred - y) * ((60*x)/(w**2 * np.exp(x/w)))
```

```python
# fit to training set with gradient descent\
alpha = .1 # learning rate
n = len(players_avg_val["rank"]) # number of 

kf = KFold(n_splits=5, shuffle=True, random_state=1)
kfold_results = {"w":[], "mae":[], "rmse":[], "r2":[]}

for idx, (train_ind, val_ind) in enumerate(kf.split(X_train_ranks, y_train)):
    w = 30 # 30 looks like a good starting point from visual inspection
    
    X_train_kf, y_train_kf = X_train_ranks[train_ind], y_train[train_ind]
    X_val_kf, y_val_kf = X_train_ranks[val_ind], y_train[val_ind]
    
    for epoch in range(20):
        for rank, avg_value in zip(X_train_kf, y_train_kf):
            w -= alpha * d_mse_wrt_w(rank, w, avg_value)
    
    preds = [f(rank, w) for rank in X_val_kf]
    
    kfold_results["w"].append(w)
    kfold_results["mae"].append(metrics.mean_absolute_error(y_val_kf, preds))
    kfold_results["rmse"].append(np.sqrt(metrics.mean_squared_error(y_val_kf, preds)))
    kfold_results["r2"].append(metrics.r2_score(y_val_kf, preds))
```

```python
w = np.mean(kfold_results["w"])
print("average w from cross validations:", w.round(3))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_train_ranks, y_train)
plt.title("Average Value vs Rank (Training Data)", fontsize=16)
plt.xlabel("Player Rank (previous year)", fontsize=14)
plt.ylabel("Player Average Draft Value", fontsize=14)
plt.plot(np.linspace(0,250, num=500), [f(x, w) for x in np.linspace(0,250, num=500)], color="g");
```

Across five fold cross validation, the average *w* parameter was 32.7 after 20 epochs of training. The curve above shows the inverse exponential function with this value. Now lets make predictions on the test set based on this model.

```python
preds = [f(rank, w) for rank in X_test_ranks]

print("MAE: {}\nR^2: {}\nRMSE: {}".format(metrics.mean_absolute_error(y_test, preds).round(3),
                                metrics.r2_score(y_test, preds).round(3),
                                np.sqrt(metrics.mean_squared_error(y_test, preds).round(3))))
```

Not too bad! The **inverse exponential** model **fitted only on a single feature** (rank) can predict the average draft value of a player with an **average error of about 5.6** fantasy dollars, which is an R^2 value of about 0.74. Lets see if a more advanced model can improve on this.

## Modeling 2 - more complex model
NGBoost is a probabilistic boosting model. I chose to use a boosting model because the target variable isn't linear. I chose NGBoost in particular because it also allows you to visualize the uncertainty of each prediction it makes.

```python
# fit to normal distribution
ngb = NGBRegressor(Dist=Normal, Score=CRPS).fit(X_train_scaled, y_train)
print("train set r^2: {}".format(metrics.r2_score(y_train, ngb.predict(X_train_scaled)).round(3)))
print("test set r^2: {}".format(metrics.r2_score(y_test, ngb.predict(X_test_scaled)).round(3)))
```

```python
# MAE: this is the average error, in fantasy dollars, for the test set predictions
print("MAE: {}\nRMSE: {}".format(metrics.mean_absolute_error(y_test, ngb.predict(X_test_scaled)),
np.sqrt(metrics.mean_squared_error(y_test, ngb.predict(X_test_scaled)))))
```

The model overfits a little bit on the training data, but not too much. An MAE of about 6.8 fantasy dollars is pretty good in this context; top value players go for more than 40 dollars so 7 dollars off still gives you a good idea of what the price might be.

However, **the simple inverse exponential model actually does a better job!** In this case the extra complexity didn't add value, possibly due to NGBoost overfitting on the small training set.

### NGBoost prediction distributions
Get a distribution for each prediction using NGBoost.

```python
pred_dists = ngb.pred_dist(X_test_scaled)
```

```python
def print_pred_dist(pred_dists, X_test_idx, y_true, X_test_names):
    """
    prints a distribution for a prediction
    
    arg pred_dists: a list of all the prediction distributions (from e.g.: ngb.pred_dist(X_test))
    """
    assert X_test_idx in range(len(pred_dists))
    
    xs = np.linspace(0, 60, num=100)
    ys = pred_dists[X_test_idx].pdf(xs).T
    
    plt.figure(figsize=(10,5))
    plt.plot(xs, ys)
    plt.axvline(pred_dists[X_test_idx].params["loc"], linestyle="--", label="Expected Price") # expected value
    plt.axvline(y_true[X_test_idx], color="r", label="True Price")
    plt.title("{}".format(X_test_names.values[X_test_idx]))
    plt.xlabel("Price (Fantasy $)")
    plt.ylabel("Probability Density")
    plt.legend()
    
    print("Expected Price: ${}".format((pred_dists[X_test_idx].params["loc"]).round(2)))
    print("True Price: ${}".format((y_test[X_test_idx]).round(2)))
```

```python
print_pred_dist(pred_dists, 11, y_test, X_test_names)
```

## Analysis
Fitting a inverse exponential curve to the data was best for predicting the average fantasy dollar value of players. This simple method probably outperformed the complex NGBoost model because of the small amount of training data, which caused NGBoost to overfit. 

However, both of these methods still had a decent amount of error, about 6 fantasy dollars off on average. Predicting draft values is a hard problem, mainly because the value isn't entirely dependent on stats; it's heavily influenced by randomness in the draft order, how well people budget their money, etc. In other words, many human aspects come into play which are hard to model. The predictions do not take into account rising stars, trades, etc., which need to be factored in by the model user. The predictions are **only accurate for players who are expected to have the same worth as the previous year**.

Nevertheless, these models give a good idea of what a player will be worth, helping to budget for the draft.
